# Stacks

Methods/Patterns: Stack

*Easy* - Valid Parentheses - Only append the opening parantheses. Once a closing parantheses arrives, check if the top of the stack contains the matching opening parantheses, if not then instantly return false as that is not a valid expression. Make sure to check at the end that the stack is empty as it must be to be able to return True.

Learnings:
- Instead of checking if a character is in the keys of the dictionary. Just check if the character is in the dictionaries. **Dictionary membership already checks keys**.

Time & Space Complexity
- Using 
  - Time complexity: O(n) -> we iterate through the string just once
  - Space complexity: O(n) -> worst case scenario is if the strings has only opening brackets so we append them all to the stack
