# Two Pointers

Methods/Patterns: Two pointers

*Easy* - Is Palindrome -Use two pointers, one on the left, another on the right to check if the string reads the same in both directions by moving the pointers.

Learnings:
- **.isalnum()** checks if the character is alpha numeric
- While loops are very useful in this scenario where we want to skip and char that is not alphanumeric while also ensure that the pointers dont cross over. So while the char is alphanumeric, move the pointer.
- Try not to re-build a string, you can do the .lower() on the if statement instead of checking against an already filtered string. This saves time.
