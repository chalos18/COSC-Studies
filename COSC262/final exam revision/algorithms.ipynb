Quiz 1 Q.4 When doing a recursion where you add things to a list it is best if you assign it to a variable and then .insert() what you want to store in the list. Don't return the recursive call, instead return where the recursive call is stored, e.g we are returning remaining instead of returning remaining.insert(...)

```python
def odds(data):
    if len(data) == 0:
        return []
    else:
        num = data[0]
        remaining = odds(data[1:])
        if num % 2 != 0:
            remaining.insert(0, num)
        return remaining


print(odds([0, 1, 12, 13, 14, 9, -11, -20]))
```

Care when doing recursion where you can get a None. make sure that the first if is actually a None and not a number

If the list is not empty and the first element is not the value, it calls find recursively on the rest of the list (data[1:]).

```python
def find(data, value):
    if len(data) == 0:
        return None
    elif data[0] == value:
        return 0
    else:
        first_occurence = find(data[1:], value)
        if first_occurence is None:
            return None
        else:
            return first_occurence + 1


print(find(["hi", "there", "you", "there"], "there"))
print(find([10, 20, 30], 0))
```

Quiz 2 Q. 3 When doing recursion where you iterate using an index parameter, make sure your recursive case is correct, most cases you will use the below pattern of checking if the index has gone past the len of the data

```python
def dumbo_func(data, index=0):
    """Takes a list of numbers and does weird stuff with it"""
    if index >= len(data):
        return 0
    else:
        if (data[index] // 100) % 3 != 0:
            return 1 + dumbo_func(data, index + 1)
        else:
            return dumbo_func(data, index + 1)


data = [677, 90, 785, 875, 7, 90393, 10707]
print(dumbo_func(data))
```

When you have to write something like two for loops, e.g getting the permutations of two lists then you have to build separate recursive functions for each loop. E.g if there is 2 loops, build a function for the inner loop and another for the outer loop.

```python
def all_pairs_inner(list1_el, list2, start_index=0):
    if start_index >= len(list2):
        return []
    else:
        iterator = list2[start_index]
        return [(list1_el, iterator)] + all_pairs_inner(
            list1_el, list2, start_index + 1
        )


def all_pairs(list1, list2, start_index=0):
    if start_index >= len(list1):
        return []
    else:
        iterator = list1[start_index]
        return (all_pairs_inner(iterator, list2)) + all_pairs(
            list1, list2, start_index + 1
        )

print(all_pairs([1, 2], [10, 20, 30]))
```
